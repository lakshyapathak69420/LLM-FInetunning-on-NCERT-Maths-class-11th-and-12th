```python
def format(vec):
    x = vec[0]
    y = vec[1]
    z = f"<s><|user|>\n{x}</s><|assistant|>{y}</s>"
    return z
```

```python
formatted_prompt = data.apply(format, axis=1)
```

```python
tokeniser.add_bos_token = "<s>"
tokeniser.add_eos_token = "</s>"
```

```python
def tokenise(formatted_prompt, tokeniser):
    tokenised_prompt = []
    for p in range(len(formatted_prompt)):

        string_prompt = str(formatted_prompt.iloc[p])
        output_prompt = tokeniser(string_prompt)
        tokenised_prompt.append(output_prompt)
    return tokenised_prompt
```

```python
tokenised_promot = tokenise(formatted_prompt, tokeniser)
```

```python
tokenised_promot=pd.DataFrame(tokenised_promot)
```
