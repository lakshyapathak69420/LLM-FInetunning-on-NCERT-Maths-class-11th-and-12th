```python
import pymupdf as pmp
```

```python
maths11 = "/home/black_tree/Desktop/NLP project /data/class 11th maths"
data = "/home/black_tree/Desktop/NLP project /data"
maths12 = "/home/black_tree/Desktop/NLP project /data/class 12th maths"
```

```python
import os 
```

```python
maths11thpdf = os.listdir(maths11)
maths12thpdf = os.listdir(maths12)
```

```python
maths11thpdf
```

```python
for i , chapter in enumerate(maths11thpdf):
    maths11thpdf[i] = os.path.join(maths11 , chapter)
```

```python
for i , chapter in enumerate(maths12thpdf):
    maths12thpdf[i] = os.path.join(maths12, chapter)
```

```python
os.chdir(data)
os.makedirs(os.path.join(data, "all of the chapters"))
os.chdir(os.path.join(data, "all of the chapters"))
for i , chapter in enumerate(maths11thpdf):
    docs = pmp.open(chapter)
    text = chr(12).join([page.get_text(sort = True) for page in docs])
    fname = f"chapter {i}"
    with open(fname , "w") as f:
        f.write(text)
```

```python
for i , chapter in enumerate(maths12thpdf):
    docs = pmp.open(chapter)
    text = chr(12).join([page.get_text(sort = True) for page in docs])
    fname = f"chapter {i + 14}"
    with open(fname , "w") as f:
        f.write(text)
```
